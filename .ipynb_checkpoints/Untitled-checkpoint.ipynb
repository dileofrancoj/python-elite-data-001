```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./alcohol.csv')
```

```python
# Limpieza de `scholarship_2023`
df['scholarship_2023'] = df['scholarship_2023'].astype(str).str.strip()  # Convertir todo a string y limpiar espacios
df['scholarship_2023'] = df['scholarship_2023'].map({'No': 0, 'Yes (NSFAS, etc...)': 1})

# Eliminar filas con valores no v√°lidos en scholarship_2023
# reemplazar por la moda / eliminar los invalidos
df = df.dropna(subset=['scholarship_2023'])
```

```python
df.dropna(subset=['scholarship_2023', 'gpa_2023'], inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='scholarship_2023', y='gpa_2023', data=df, palette='muted')
plt.title("Distribuci√≥n del GPA seg√∫n Estado de Beca", fontsize=14)
plt.xticks([0, 1], ['Sin Beca', 'Con Beca'], fontsize=12)
plt.xlabel("Estado de Beca", fontsize=12)
plt.ylabel("GPA 2023", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='gpa_2023', hue='scholarship_2023', bins=20, kde=True, 
             palette=['blue', 'orange'], edgecolor='black', alpha=0.7)

plt.title("Histograma de GPA 2023 seg√∫n Estado de Beca", fontsize=14)
plt.xlabel("GPA 2023", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(['Sin Beca', 'Con Beca'], title='Estado de Beca', fontsize=12)
plt.show()
```

### Correlaci√≥n

La correlaci√≥n estad√≠stica mide la relaci√≥n entre dos variables y c√≥mo cambian juntas. Es una herramienta que permite identificar si existe una conexi√≥n lineal entre las variables y, si es as√≠, qu√© tan fuerte y en qu√© direcci√≥n es esa conexi√≥n.

- Positiva: Ambas variables aumentan juntas. Un ejemplo podr√≠a ser: A mayor ingreso, mayor gasto
- Negativa: Hay una relaci√≥n inversa. Si una aumenta, la otra disminuye. Mayor velocidad, menor tiempo de viaje
- Nula: No hay relacion entre esas variables

0.7 ‚â§ |r| < 1: Correlaci√≥n fuerte.

0.2 ‚â§ |r| < 0.7: Correlaci√≥n moderada.

0 ‚â§ |r| < 0.2: Correlaci√≥n d√©bil.

### Tipos de coeficientes:

- Pearson: Eval√∫a la relaci√≥n lineal entre dos variables continuas
- Spearman: Eval√∫a la relaci√≥n entre variables ordinales
- Kendall: Similar a Spearman pero es m√°s robusto ante peque√±os tama√±os de muestra

```python
numeric_df = df.select_dtypes(include=['number'])

correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlaciones", fontsize=16)
plt.show()
```

### Kurtosis

Curtosis es una medida estad√≠stica que describe la forma de la distribuci√≥n de los datos, especialmente enfoc√°ndose en la "agudeza" o "aplanamiento" de sus colas. En otras palabras, se refiere a c√≥mo est√°n distribuidos los datos en torno a la media, considerando especialmente las colas (los extremos) de la distribuci√≥n.

### Tipos de Curtosis:

- Leptoc√∫rtica: Distribuci√≥n muy "aguda", con colas pesadas, es decir, valores extremos muy presentes. (Curtosis > 0)
- Mesoc√∫rtica: Distribuci√≥n similar a una distribuci√≥n normal, ni muy aplanada ni muy aguda. (Curtosis ‚âà 0)
- Platic√∫rtica: Distribuci√≥n "aplanada", con colas ligeras y menos valores extremos. (Curtosis < 0)

```python
gpa_kurtosis = df['gpa_2023'].kurtosis()

print(f"Curtosis de GPA 2023: {gpa_kurtosis}")
```

### ANOVA

ANOVA (An√°lisis de Varianza) es una t√©cnica estad√≠stica utilizada para comparar las diferencias entre las medias de tres o m√°s grupos. Su objetivo principal es determinar si hay diferencias significativas entre las medias de los grupos, bas√°ndose en la variabilidad entre las medias y la variabilidad dentro de los grupos.

En otras palabras, ANOVA analiza si las diferencias observadas en las variables dependientes (como GPA, ingresos, etc.) se deben a factores espec√≠ficos, como categor√≠as o grupos independientes.



```python
# Comparaci√≥n de GPA entre diferentes g√©neros
sns.boxplot(x='gender', y='gpa_2023', data=df)
plt.title("Comparaci√≥n de GPA por G√©nero", fontsize=14)
plt.show()

# ANOVA para ver diferencias significativas entre faculdades
from scipy.stats import f_oneway

# ANOVA entre 'faculty' y 'gpa_2023'
anova_result = f_oneway(
    df[df['faculty'] == 'Arts & Social Sciences']['gpa_2023'],
    df[df['faculty'] == 'Engineering']['gpa_2023'],
    df[df['faculty'] == 'Economic & Management Sciences']['gpa_2023']
)
print(f"ANOVA Result: F-statistic={anova_result.statistic}, p-value={anova_result.pvalue}")
```

### Interpretaci√≥n del Resultado:
F-statistic: Indica la relaci√≥n entre las variables dentro de los grupos. En este caso, un valor bajo sugiere que las diferencias en GPA entre las diferentes facultades son peque√±as.

p-value: Muy alto (cercano a 1), lo que indica que no hay diferencias estad√≠sticamente significativas entre las facultades en cuanto a GPA 2023.

Conclusi√≥n:
Este resultado sugiere que, estad√≠sticamente, las diferencias en GPA entre las diferentes facultades podr√≠an ser consideradas insignificantes. Esto puede deberse a la diversidad de facultades y sus contextos √∫nicos que afectan los GPA, pero en t√©rminos generales, los datos no muestran una diferencia significativa entre ellas.
